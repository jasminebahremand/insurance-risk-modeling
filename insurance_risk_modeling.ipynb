# Data Cleaning

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Define the y cols
df_train['CS'] = (df_train['X.16'] > 0).astype(int)  #ty x15 or 16
df_train['LC'] = 0.0
df_train['HALC'] = 0.0
nonzero_mask = df_train['X.16'] > 0
df_train.loc[nonzero_mask, 'LC'] = df_train.loc[nonzero_mask, 'X.15'] / df_train.loc[nonzero_mask, 'X.16']
df_train.loc[nonzero_mask, 'HALC'] = df_train.loc[nonzero_mask, 'LC'] * df_train.loc[nonzero_mask, 'X.18']


#Dropping Columns
cols_to_drop = ['X.1', 'X.15', 'X.16', 'X.17', 'X.18']
df_train.drop(columns=[col for col in cols_to_drop if col in df_train.columns], inplace=True)

# Convert date columns to datetime
date_cols = ['X.2', 'X.3', 'X.4', 'X.5', 'X.6']
for col in date_cols:
    df_train[col] = pd.to_datetime(df_train[col], errors='coerce', dayfirst=True)


# Feature engineering
df_train['vehicle_age'] = 2020 - df_train['X.22']                    # from registration year
df_train['age'] = 2020 - df_train['X.5'].dt.year                     # from birthdate
df_train['driving_experience'] = 2020 - df_train['X.6'].dt.year     # from license issue date
df_train['policy_duration_days'] = (df_train['X.4'] - df_train['X.2']).dt.days
df_train['time_since_last_renewal_days'] = (df_train['X.3'] - df_train['X.2']).dt.days

# Drop raw date columns from both train and test
df_train.drop(columns=[col for col in date_cols if col in df_train.columns], inplace=True)

# One-hot encoding
categorical_cols = ['X.7', 'X.13', 'X.19', 'X.20', 'X.21', 'X.27']
df_train = pd.get_dummies(df_train, columns=categorical_cols, drop_first=True)



```

# Task 1

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import TweedieRegressor
from sklearn.metrics import mean_squared_error

df=df_train
# Drop raw dates and leakage columns
drop_cols = ["LC", "HALC", "CS"]

# Filter out extreme outliers
q_lc = df["LC"].quantile([0.001, 0.999])
q_halc = df["HALC"].quantile([0.001, 0.999])
#q_lc = df["LC"].quantile([0.01, 0.99])
#q_halc = df["HALC"].quantile([0.01, 0.99])
df_filtered = df[(df["LC"] >= q_lc.iloc[0]) & (df["LC"] <= q_lc.iloc[1]) &
               (df["HALC"] >= q_halc.iloc[0]) & (df["HALC"] <= q_halc.iloc[1])]

# Prepare features and targets
X = df_filtered.drop(columns=drop_cols)
#X = pd.get_dummies(X, columns=["X.27"], drop_first=True)
y_lc = df_filtered["LC"]
y_halc = df_filtered["HALC"]

# Split and scale
X_train_lc, X_val_lc, y_train_lc, y_val_lc = train_test_split(X, y_lc, test_size=0.2, random_state=42)
X_train_halc, X_val_halc, y_train_halc, y_val_halc = train_test_split(X, y_halc, test_size=0.2, random_state=42)

scaler_lc = StandardScaler().fit(X_train_lc)
X_train_lc_scaled = scaler_lc.transform(X_train_lc)
X_val_lc_scaled = scaler_lc.transform(X_val_lc)

scaler_halc = StandardScaler().fit(X_train_halc)
X_train_halc_scaled = scaler_halc.transform(X_train_halc)
X_val_halc_scaled = scaler_halc.transform(X_val_halc)

# Fit Tweedie Regressors
glm_lc = TweedieRegressor(power=1.5, link='log', alpha=0.0, max_iter=1000)
glm_lc.fit(X_train_lc_scaled, y_train_lc)
y_pred_lc = glm_lc.predict(X_val_lc_scaled)
mse_lc = mean_squared_error(y_val_lc, y_pred_lc)
rmse_lc = np.sqrt(mse_lc)

glm_halc = TweedieRegressor(power=1.5, link='log', alpha=0.0, max_iter=1000)
glm_halc.fit(X_train_halc_scaled, y_train_halc)
y_pred_halc = glm_halc.predict(X_val_halc_scaled)
mse_halc = mean_squared_error(y_val_halc, y_pred_halc)
rmse_halc = np.sqrt(mse_halc)

print(f"LC RMSE: {rmse_lc:.2f}")
print(f"HALC RMSE: {rmse_halc:.2f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_lc, y_train_lc)

y_pred_rf = rf.predict(X_val_lc)
mse_rf = mean_squared_error(y_val_lc, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)

print("Random Forest - LC")
print(f"MSE: {mse_rf:.2f}")
print(f"RMSE: {rmse_rf:.2f}")

# Reuse same features
rf_halc = RandomForestRegressor(n_estimators=100, random_state=42)
rf_halc.fit(X_train_halc, y_train_halc)

y_pred_rf_halc = rf_halc.predict(X_val_halc)
mse_rf_halc = mean_squared_error(y_val_halc, y_pred_rf_halc)
rmse_rf_halc = np.sqrt(mse_rf_halc)

print("Random Forest - HALC")
print(f"MSE: {mse_rf_halc:.2f}")
print(f"RMSE: {rmse_rf_halc:.2f}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
gb.fit(X_train_lc, y_train_lc)

y_pred_gb = gb.predict(X_val_lc)
mse_gb = mean_squared_error(y_val_lc, y_pred_gb)
rmse_gb = np.sqrt(mse_gb)

print("Gradient Boosting - LC")
print(f"MSE: {mse_gb:.2f}")
print(f"RMSE: {rmse_gb:.2f}")

gb_halc = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
gb_halc.fit(X_train_halc, y_train_halc)

y_pred_gb_halc = gb_halc.predict(X_val_halc)
mse_gb_halc = mean_squared_error(y_val_halc, y_pred_gb_halc)
rmse_gb_halc = np.sqrt(mse_gb_halc)

print("Gradient Boosting - HALC")
print(f"MSE: {mse_gb_halc:.2f}")
print(f"RMSE: {rmse_gb_halc:.2f}")
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import TweedieRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Models
models = {
    "GLM (Tweedie)": TweedieRegressor(power=1.5, link='log', alpha=0.0, max_iter=1000),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "LightGBM": LGBMRegressor(n_estimators=100, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=100, random_state=42, verbosity=0),
    "Neural Network": MLPRegressor(hidden_layer_sizes=(64, 32), max_iter=300, random_state=42)
}

# Store results
cv_results = []

# Loop through models
for name, model in models.items():
    print(f"Running CV for: {name}")

    # Scale only for GLM and MLP
    if name in ['GLM (Tweedie)', 'Neural Network']:
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
    else:
        X_scaled = X.values  # no scaling for tree models

    # LC
    mse_lc_list, rmse_lc_list = [], []
    for train_idx, val_idx in kf.split(X_scaled):
        X_train, X_val = X_scaled[train_idx], X_scaled[val_idx]
        y_train, y_val = y_lc.iloc[train_idx], y_lc.iloc[val_idx]

        model.fit(X_train, y_train)
        preds = model.predict(X_val)

        mse = mean_squared_error(y_val, preds)
        rmse = np.sqrt(mse)
        mse_lc_list.append(mse)
        rmse_lc_list.append(rmse)

    # HALC
    mse_halc_list, rmse_halc_list = [], []
    for train_idx, val_idx in kf.split(X_scaled):
        X_train, X_val = X_scaled[train_idx], X_scaled[val_idx]
        y_train, y_val = y_halc.iloc[train_idx], y_halc.iloc[val_idx]

        model.fit(X_train, y_train)
        preds = model.predict(X_val)

        mse = mean_squared_error(y_val, preds)
        rmse = np.sqrt(mse)
        mse_halc_list.append(mse)
        rmse_halc_list.append(rmse)

    # Store results
    cv_results.append({
        "Model": name,
        "LC MSE": np.mean(mse_lc_list),
        "LC RMSE": np.mean(rmse_lc_list),
        "HALC MSE": np.mean(mse_halc_list),
        "HALC RMSE": np.mean(rmse_halc_list)
    })

# Create summary DataFrame
cv_summary = pd.DataFrame(cv_results).round(2)
cv_summary = cv_summary.sort_values(by="HALC RMSE")

# Highlight best RMSEs
def highlight_best(s):
    is_min = s == s.min()
    return ['background-color: lightgreen' if v else '' for v in is_min]

styled_cv_summary = cv_summary.style.apply(highlight_best, subset=['LC RMSE', 'HALC RMSE'])

styled_cv_summary
```

```python
# Check the distributuion 0/1
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def check_class_balance(y, plot=True, threshold=0.4):
    y_series = pd.Series(y, name="label")
    counts = y_series.value_counts().sort_index()
    props  = counts / counts.sum()

    print("Class counts")
    print(counts, "\n")
    print("Class proportions")
    print(props.apply(lambda p: f"{p:.2%}"), "\n")

    minority_prop = props.min()
    if minority_prop < threshold or minority_prop > 1 - threshold:
        print(f"  Data appears imbalanced "
              f"(minority class = {minority_prop:.2%} < {threshold:.0%}).")
    else:
        print(" Data looks reasonably balanced.")

    if plot:
        sns.barplot(x=counts.index.astype(str), y=counts.values, palette="Set2")
        plt.ylabel("Number of samples")
        plt.xlabel("Class label")
        plt.title("Class distribution")
        plt.show()

    return counts, props , counts.sum()

counts, proportions, total_count = check_class_balance(y_cs_train, plot=True, threshold=0.4)
```
